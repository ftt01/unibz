```python
### evaluate the monthly average
## aggredate the data
gs_prec_jan = prec_data.loc[(prec_data.index.month==1)]
gs_prec_feb = prec_data.loc[(prec_data.index.month==2)]
gs_prec_mar = prec_data.loc[(prec_data.index.month==3)]
gs_prec_apr = prec_data.loc[(prec_data.index.month==4)]
gs_prec_may = prec_data.loc[(prec_data.index.month==5)]
gs_prec_jun = prec_data.loc[(prec_data.index.month==6)]
gs_prec_jul = prec_data.loc[(prec_data.index.month==7)]
gs_prec_aug = prec_data.loc[(prec_data.index.month==8)]
gs_prec_sep = prec_data.loc[(prec_data.index.month==9)]
gs_prec_oct = prec_data.loc[(prec_data.index.month==10)]
gs_prec_nov = prec_data.loc[(prec_data.index.month==11)]
gs_prec_dec = prec_data.loc[(prec_data.index.month==12)]

## evaluate the mean, dropping stations with all NaN
gs_prec_jan_mean = gs_prec_jan.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_feb_mean = gs_prec_feb.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_mar_mean = gs_prec_mar.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_apr_mean = gs_prec_apr.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_may_mean = gs_prec_may.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_jun_mean = gs_prec_jun.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_jul_mean = gs_prec_jul.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_aug_mean = gs_prec_aug.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_sep_mean = gs_prec_sep.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_oct_mean = gs_prec_oct.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_nov_mean = gs_prec_nov.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()
gs_prec_dec_mean = gs_prec_dec.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nansum(x)).dropna(how='all').mean(axis=1).mean()

total_prec_yearly_mean = np.sum([
    gs_prec_jan_mean,gs_prec_feb_mean,gs_prec_mar_mean,
    gs_prec_apr_mean,gs_prec_may_mean,gs_prec_jun_mean,
    gs_prec_jul_mean,gs_prec_aug_mean,gs_prec_sep_mean,
    gs_prec_oct_mean,gs_prec_nov_mean,gs_prec_dec_mean])

print(total_prec_yearly_mean)
```

```python
### evaluate the monthly average
## aggredate the data
gs_temp_jan = temp_data.loc[(temp_data.index.month==1)]
gs_temp_feb = temp_data.loc[(temp_data.index.month==2)]
gs_temp_mar = temp_data.loc[(temp_data.index.month==3)]
gs_temp_apr = temp_data.loc[(temp_data.index.month==4)]
gs_temp_may = temp_data.loc[(temp_data.index.month==5)]
gs_temp_jun = temp_data.loc[(temp_data.index.month==6)]
gs_temp_jul = temp_data.loc[(temp_data.index.month==7)]
gs_temp_aug = temp_data.loc[(temp_data.index.month==8)]
gs_temp_sep = temp_data.loc[(temp_data.index.month==9)]
gs_temp_oct = temp_data.loc[(temp_data.index.month==10)]
gs_temp_nov = temp_data.loc[(temp_data.index.month==11)]
gs_temp_dec = temp_data.loc[(temp_data.index.month==12)]


## evaluate the mean, dropping stations with all NaN
gs_temp_jan_mean = gs_temp_jan.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_feb_mean = gs_temp_feb.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_mar_mean = gs_temp_mar.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_apr_mean = gs_temp_apr.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_may_mean = gs_temp_may.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_jun_mean = gs_temp_jun.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_jul_mean = gs_temp_jul.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_aug_mean = gs_temp_aug.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_sep_mean = gs_temp_sep.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_oct_mean = gs_temp_oct.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_nov_mean = gs_temp_nov.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()
gs_temp_dec_mean = gs_temp_dec.resample("MS").agg(lambda x: np.nan if np.isnan(x).all() else np.nanmean(x)).dropna(how='all').mean(axis=1).mean()

gs_temp_w_mean=np.mean([gs_temp_dec_mean,gs_temp_jan_mean,gs_temp_feb_mean])
gs_temp_sp_mean=np.mean([gs_temp_mar_mean,gs_temp_apr_mean,gs_temp_may_mean])
gs_temp_su_mean=np.mean([gs_temp_jun_mean,gs_temp_jul_mean,gs_temp_aug_mean])
gs_temp_a_mean=np.mean([gs_temp_sep_mean,gs_temp_oct_mean,gs_temp_nov_mean])

print("T mean")
print(gs_temp_w_mean)
print(gs_temp_sp_mean)
print(gs_temp_su_mean)
print(gs_temp_a_mean)

gs_temp_w_min=np.min([gs_temp_dec_mean,gs_temp_jan_mean,gs_temp_feb_mean])
gs_temp_sp_min=np.min([gs_temp_mar_mean,gs_temp_apr_mean,gs_temp_may_mean])
gs_temp_su_min=np.min([gs_temp_jun_mean,gs_temp_jul_mean,gs_temp_aug_mean])
gs_temp_a_min=np.min([gs_temp_sep_mean,gs_temp_oct_mean,gs_temp_nov_mean])

print("T min")
print(gs_temp_w_min)
print(gs_temp_sp_min)
print(gs_temp_su_min)
print(gs_temp_a_min)

gs_temp_w_max=np.max([gs_temp_dec_mean,gs_temp_jan_mean,gs_temp_feb_mean])
gs_temp_sp_max=np.max([gs_temp_mar_mean,gs_temp_apr_mean,gs_temp_may_mean])
gs_temp_su_max=np.max([gs_temp_jun_mean,gs_temp_jul_mean,gs_temp_aug_mean])
gs_temp_a_max=np.max([gs_temp_sep_mean,gs_temp_oct_mean,gs_temp_nov_mean])

print("T max")
print(gs_temp_w_max)
print(gs_temp_sp_max)
print(gs_temp_su_max)
print(gs_temp_a_max)
```
